```python
katok = []
num = []
a=0

def del_data(position) :
    klen = len(katok) 
    katok[position] = None 
    for i in range(position+1, klen):
        katok[i-1] = katok[i] 
        katok[i] = None
    del(katok[klen-1]) 

def add_data(friend):
    global a
    a = 0
    c = [i for i in friend if i.isdigit()] 
    f = int(''.join(c)) 
    num.append(f) 
    for i in num: 
        if i > f: 
            a+=1
    katok.insert(a,friend)

if __name__ == "__main__":
    select = -1

    while(select!=3):
        select = int(input("선택하세요(1:추가, 2:삭제, 3:종료)--> "))

        if select == 1:
            data = input("추가할 데이터--> ")
            add_data(data)
            print(katok)

        elif select == 2:
            data = int(input("삭제할 순서--> "))
            if data<0 or data>len(katok):
                print("삭제할 데이터가 존재하지 않습니다.")
                continue 
            del_data(data-1)
            print(katok)

        elif select == 3:
            exit

        else:
            print("1~3까지의 숫자만 입력하세요.")
            continue
```
